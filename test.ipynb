```python
import vt
import os
import asyncio
from pprint import pprint
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Access the API key using os.getenv
vt_api_key = os.getenv("VT_API_KEY")

if vt_api_key is None:
    raise ValueError("VT_API_KEY not found in environment variables")

# Any File/Hash IOC can be used for a VT lookup
ioc = "3e55bf8ecaeec65871e6fca4cb2d4ff2586f83a20c12977858348492d2d0dec4"

async def get_vt_result(ioc):
    async with vt.Client(vt_api_key) as client:
        result = await client.get_object_async(f"/files/{ioc}")
        return result

result = await get_vt_result(ioc)
pprint(result.last_analysis_stats) # Print the last analysis results
```
